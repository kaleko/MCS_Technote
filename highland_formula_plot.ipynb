```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
seglen = 10

def highland_sigma(p):
    m = 0.106
    beta = np.sqrt( 1 - ((m*m)/(p*p + m*m)) )
    red_length = float(seglen)/14.
    return ( 12.265 / (p*beta) ) * ( 1.0 + 0.038 * np.log10( red_length ) ) * np.sqrt( red_length )
```

```python
ps = np.linspace(0,6,100)
sigs = highland_sigma(ps)
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('Muon Momentum, $p$ [GeV]',fontsize=16)
plt.ylabel('Highland Predicted $\sigma_o$ [mrad]',fontsize=16)
plt.title('Visualizing the Highland Formula',fontsize=16)
plt.plot(ps,sigs,'r-',linewidth=5)
plt.grid(True,'both')
plt.yscale('log')
```
